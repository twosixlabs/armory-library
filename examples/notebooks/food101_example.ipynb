## Imports

```python
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from matplotlib import pyplot as plt
ROOT = "/home/chris/cache"
```

## Load Pokemon Dataset From HuggingFace into Jatic Dataset

```python
train = datasets.Food101(
    ROOT,
    split="train",
    transform=transforms.Compose(
        [
            transforms.Resize(size=(512, 512)),
        ]
    ),
    download=False,
)
dataloader = DataLoader(train, batch_size=1, num_workers=4)
```

## Take out 4 samples

```python
food_labels_to_show = [1,1000,3000,40000]
food_examples = [train[i][0] for i in food_labels_to_show]
```

## Displaying Images

```python
plt.figure()
for pokemon in food_examples:
    plt.imshow(pokemon)
    plt.show(block=True)
```
