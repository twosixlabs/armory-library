## Imports

```python
from jatic_toolbox import load_dataset as load_jatic_dataset
from charmory.experimental.examples import show_samples
from charmory.utils import PILtoNumpy
from torchvision import transforms as T
from pathlib import Path
ROOT = str(Path.home()) + "/cache"
```

## Load Food101 Dataset From Torchvision into Jatic Dataset

```python
train_dataset = load_jatic_dataset(
        provider="torchvision",
        dataset_name="Food101",
        task="image-classification",
        split="train",
        root=ROOT,
        download=True,
        transform=T.Compose(
            [
                T.Resize(size=(512, 512)),
                PILtoNumpy(),
            ]
        ),
    )
```

## Displaying Images

```python
show_samples(train_dataset)
```
